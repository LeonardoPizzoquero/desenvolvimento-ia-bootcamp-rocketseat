```python
import pandas as pd

dados_vendas = {
    'mes': ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro'],
    'vendas': [2000, 2200, 2300, 2500, 2600, 2700, 2800, 2900, 3000, 3100, 3200, 3300]
}

df = pd.DataFrame.from_dict(dados_vendas)
```

```python
df.shape
```

```python
df.head(12)
```

```python
# Transformar o mês de 1 a 12
df['mes_num'] = range(1, 13)
```

```python
from sklearn.linear_model import LinearRegression

X = df[['mes_num']][df['mes_num'] < 12]
y = df['vendas'][df['mes_num'] < 12]

X_test = df[['mes_num']][df['mes_num'] == 12]

model = LinearRegression()

model.fit(X, y)

y_pred = model.predict(X_test)

y_pred
```

```python
model.coef_, model.intercept_
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot de Scatter com distribuição
plt.figure(figsize=(10, 6))
sns.scatterplot(x='mes_num', y='vendas', data=df)
plt.plot(df['mes_num'], model.predict(df[['mes_num']]), color='red')
plt.title('Scatter Plot com Distribuição de Vendas Mensais')
plt.xlabel('Mês')
plt.ylabel('Vendas')
plt.grid(True)
plt.show()

# Histograma de todas as variáveis
df.hist(bins=10, figsize=(10, 6), grid=True)
plt.suptitle('Histogramas de Vendas e Meses')
plt.show()
```
